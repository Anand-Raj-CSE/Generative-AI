Corpus - paragrah , Documents - sentences , vocabulary - unique words. Tokenization is process of converting paragraph to sentences and then sentences to words and then creating a vocabulary of unique words in our corpus.

Tokenization -> process of breaking down a large paragraph into smaller units called tokens. These tokens can be words, phrases, or even sentences. Tokenization is a crucial step in natural language processing (NLP) as it allows us to analyze and manipulate text data more effectively
